## CATGORICAL PLOTS

```python
import seaborn as sns

%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
# x - categorical column
# y - numerical column
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
# estimator - average or mean or custom function
# we are passing broadcasting functions
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# counts the number of occurences
sns.countplot(x='sex', data=tips)
```

### BOX PLOTS AND VIOLIN PLOTS
- USED FOR SHOWING DISTRIBUTION OF CATEGORICAL DATA

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# hue - adding a whole new layer of categorical data
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

#### VIOLIN PLOT
- plot all the components that correspond to the actual data points
- shows KDE of the underlying distribution

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### STRIP PLOT
- WE CAN'T TELL HOW MANY DATA POINTS ARE STACKED ON TOP OF EACH OTHER

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# jitter adds additional noise
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

### SWARM PLOT

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```
