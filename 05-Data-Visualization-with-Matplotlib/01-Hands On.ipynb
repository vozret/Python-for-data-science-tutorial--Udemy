#### setting two plots

```python
# creating the figure object

fig = plt.figure()

# adding the first axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# adding the second axes
# 20% from the left corner, 50% up from the bottom,
# 40% of the canvas width, 30% of the canvas heigth
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
# moving the second figure to the botom right corner

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.1, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('LARGER PLOT')

axes2.plot(y, x)
axes2.set_title('SMALLER PLOT')
```

### FIGURE SIZE ASPECT RATIO AND DPI

```python
fig = plt.figure(figsize=(8, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

#### LEGENDS

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('y Label')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.plot(y, x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2)
ax.plot(x, x**3)
```

```python
# in creating a legend
# we have to set the 'label' argument in ax.plot()

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='x squared')
ax.plot(x, x**3, label = 'x cubed')

ax.legend()
```

```python
# we can specify the location of the legend
# we can use location string and location code

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='x squared')
ax.plot(x, x**3, label = 'x cubed')

ax.legend(loc=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='x squared')
ax.plot(x, x**3, label = 'x cubed')

ax.legend(loc=(0.1, 0.1))
```

## PART 3

#### PLOT APPEARANCE

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# we can put in the hexcode
ax.plot(x, y, color='#FF8C00') # RGB Hex Code
```

- line width and line style

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', linewidth=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# aplha argument -> transparence of the line
# linewidth == lw
ax.plot(x, y, color='purple', linewidth=3, alpha=0.5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# linestyle: '.', '-.', ':', 'steps', '--'
# linestyle == ls
ax.plot(x, y, color='purple', lw=3, linestyle='-.')
```

### CONTROL OVER AXIS APPEARANCE AND PLOT RANGE

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=2, linestyle='-')

ax.set_xlim([0, 1])
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=2, linestyle='-')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
